*Data visualization of df 

```python
#count vs category 
sns.set_style("darkgrid")
sns.countplot(x='Category',data=df,palette="Set2") 
print("Count vs Category ")
```

```python
#count vs subCategory
plt.figure(figsize=(15,9))
sns.set_style("darkgrid")
sns.countplot(x='Sub-Category',data=df,palette='Set1') 
print("Count vs subCategory ") 
```

```python
#count vs Segment
sns.set_style("darkgrid")
sns.countplot(x='Segment',data=df,palette="Set2") 
print("Count vs Segment ")
```

```python
#count vs shipmode
sns.set_style("darkgrid")
sns.countplot(x='Ship Mode',data=df,palette="Set1") 
print("Count vs shipmode")
```

```python
#count vs region
sns.set_style("darkgrid")
sns.countplot(x="Region",data=df,palette="Set2") 
print("Count vs Region")
```

```python
#Sales data 
df1=df.select_dtypes(exclude=[object])
df1.head(5)
```

```python
df1.shape
```

```python
df1.info()
```

```python
#corr matrix for sales data 
corr_matrix = df1.corr()
corr_matrix
```

```python
sns.set(rc={'figure.figsize':(9.7,5.27)})
ax=sns.heatmap(corr_matrix,vmin=-1,vmax=1,center=0,
               cmap=sns.diverging_palette(20,220,n=200),
               square = True) 
#+1 the variable is positively correlated  
#-1 the varible is negatively correlated
```

```python
#for knowing the sales density from the sales data 
plt.figure(figsize= (11,9))
sns.distplot(df.Sales, bins =150, color = 'green') 
plt.title('State density graph')
```

*State wise Sales 

```python
#State vs Sales
statewise=df.groupby(['State'])['Sales'].sum()
statewise.sort_values(ascending=False,inplace=True)
fig, axis = plt.subplots(figsize=(15,9))
plt.title('State vs Sales')
statewise.plot.bar()
```

Top 3 States with highest Sales 
    1.California 
    2.Newyork 
    3.Texas 

```python
#State vs Profit
statewise=df.groupby(['State'])['Profit'].sum()
statewise.sort_values(ascending=False,inplace=True)
fig, axis = plt.subplots(figsize=(15,10))
plt.title("State vs Profit")
statewise.plot.bar()
```

Top 3 States with highest Profit                Top 3 States with highest Loss
    1.California                                         1.Texas 
    2.Newyork                                            2.Ohio 
    3.Washington                                         3.Pennsylvania

```python
#Total Profit and sales per category
g=df.groupby('Sub-Category')['Profit','Sales'].agg(['sum'])
g.plot.bar(rot=0, title="Total Profit and sales per category",figsize=(15,10));
plt.show(block=True)
```

Top profitable product is Copiers with approximately 60000 and most sold product is phones & chairs with approximately 35000.

```python
#region wise data of sub category 
plt.figure(figsize=(15,8))
sns.countplot(x="Sub-Category", hue="Region",data=df,palette="plasma")
plt.title("Region wise Sub-Category")
plt.show()
```

People in west buy more products & people in south buy less products compared to other parts.

```python
#California State Analysis
California_data = df.where(df.State == 'California')
California_data.head(5) 
```
